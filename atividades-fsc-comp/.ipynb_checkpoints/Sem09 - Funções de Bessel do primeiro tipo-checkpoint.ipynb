### Funções de Bessel do primeiro tipo

As funções de Bessel do primeiro tipo, na sua forma integral, são dadas por

$$
J_m(x) = \frac 1 \pi \int_0^\pi \cos \left( m\theta - x \sin \theta \right)\ \mathrm{d}\theta,
$$

onde $m$ é um inteiro não negativo, e $x \geq 0$.

#### Item 1

Escreva uma função que calcule o valor de $J_m(x)$ usando o método de quadratura gaussiana (use $N=20$).

```python
import numpy as np
from scipy.special import roots_legendre

# Função integrando: cos(m * t - x * sin(t))
def f(t, x, m):
    return np.cos(m * t - x * np.sin(t))

# Função para calcular J_m(x) usando quadratura gaussiana
def J_m(x, m):
    # Configuração dos parâmetros
    N = 20  # Número de pontos de quadratura
    a = 0   # Limite inferior da integral
    b = np.pi  # Limite superior da integral
    
    # Inicialização da soma
    I = 0.0
    
    # Pontos e pesos da quadratura gaussiana
    u_k, w_k = roots_legendre(N)
    
    # Transformação para o intervalo [a, b]
    for k in range(N):
        t_k = 0.5 * (b - a) * u_k[k] + 0.5 * (b + a)
        I += w_k[k] * f(t_k, x, m)
    
    # Ajuste do fator da integral
    return I / np.pi

# Exemplo de uso:
x = 1.0  # Valor de x
m = 2    # Ordem da função de Bessel
resultado = J_m(x, m)

print(f"J_{m}({x}) = {resultado}")
```

#### Item 2

Use sua função para fazer um gráfico comparando $J_0$, $J_1$ e $J_2$ (no mesmo gráfico), com $0 \leq x \leq 20$.

```python
import matplotlib.pyplot as plt

# Função integrando: cos(m * t - x * sin(t))
def f(t, x, m):
    return np.cos(m * t - x * np.sin(t))

# Função para calcular J_m(x) usando quadratura gaussiana
def J_m(x, m):
    N = 20  # Número de pontos de quadratura
    a = 0   # Limite inferior da integral
    b = np.pi  # Limite superior da integral
    I = 0.0
    
    # Pontos e pesos da quadratura gaussiana
    u_k, w_k = roots_legendre(N)
    
    # Transformação para o intervalo [a, b]
    for k in range(N):
        t_k = 0.5 * (b - a) * u_k[k] + 0.5 * (b + a)
        I += w_k[k] * f(t_k, x, m)
    
    # Ajuste do fator da integral
    return I / np.pi

# Intervalo de x
x_values = np.linspace(0, 20, 500)

# Cálculo de J_0(x), J_1(x) e J_2(x)
J_0 = [J_m(x, 0) for x in x_values]
J_1 = [J_m(x, 1) for x in x_values]
J_2 = [J_m(x, 2) for x in x_values]

# Gráfico
plt.figure(figsize=(10, 6))
plt.plot(x_values, J_0, label=r'$J_0(x)$', linewidth=2)
plt.plot(x_values, J_1, label=r'$J_1(x)$', linestyle='--', linewidth=2)
plt.plot(x_values, J_2, label=r'$J_2(x)$', linestyle='-.', linewidth=2)
plt.xlabel('x')
plt.ylabel(r'$J_m(x)$')
plt.title('Funções de Bessel de Primeira Ordem: $J_0(x)$, $J_1(x)$, $J_2(x)$')
plt.legend()
plt.grid()
plt.show()
```
